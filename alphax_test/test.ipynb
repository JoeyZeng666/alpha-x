```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([1, 3])
y = np.array([3, 2])


plt.scatter(x, y)
```

```python
degree = 1  # 线性拟合
coeffs = np.polyfit(x, y, degree)
# slope = coeffs[0]
# intercept = coeffs[1]
slope = round(coeffs[0], 10)  # 保留10位小数
intercept = round(coeffs[1], 10)
print(slope)
print(intercept)
```

```python
fit_x = np.linspace(min(x), max(x), 100)
fit_y = np.polyval(coeffs, fit_x)
```

```python
plt.figure()
plt.scatter(x, y, label='Data points')
plt.plot(fit_x, fit_y, 'r-', label='Fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Fit using polyfit')
plt.legend()
plt.grid(True)
plt.show()
```
